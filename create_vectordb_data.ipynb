```python
for index, value in enumerate(data['data']):
    # Create a filename for each element (e.g., output/element_0.json, output/element_1.json, ...)
    #filename = f'data_json/element_{index}.json'

    final_value = {}
    final_value["tag_name"] = ""
    final_value["speakers"] = ""
    for n in important_values:

        if type(value[n]) != list:
            final_value[n.lower()] = value[n]
        else: 
            num = 0
            for index, element in enumerate(value[n]):
                for l in element:
                    if "Uid" not in l:
                        if "tags" in n:
                            if element[l]:
                                final_value["tag_name"] = final_value["tag_name"]+ element[l] + ", "
                        elif "speakers" in n:
                            if element[l]:
                                final_value["speakers"] = final_value["speakers"]+ l + "_"+str(index)+ ": " + element[l] + ", "
                        else:
                            if element[l]:
                                final_value[l.lower()+"_"+str(index)] = element[l]
                num = num+1
    texto = ""
    for h in final_value:
        if h !="sessionUid":
            texto = texto + h +" :"+ str(final_value[h]) + ", "
    final_value["summary"] = texto
    final_value["location"] = final_value.get('venuename')+", " + final_value.get('floorplanname')+", " + final_value.get('locationname')
    final_value["speakers"] = final_value["speakers"].rstrip(" ,")
    final_value["tag_name"] = final_value["tag_name"].rstrip(" ,")
    final_value["description"] = final_value["description"].rstrip(".")
    
    name_unique =final_value["thirdpartyid"].replace("-","_")+"_"+final_value["title"].replace(" ","_")
    name_unique = name_unique.replace(":","")
    name_unique = name_unique.replace("/","_")
    filename = f'agenda_reinvent/{name_unique}.txt'
    string_value = ""
    for g in final_value:
        string_value = g + ": " + final_value[g]+ "\n"+string_value

    with open(filename,"w") as file:
        file.write(string_value)
```

```python
from langchain.document_loaders import TextLoader
from pathlib import Path
```

```python
# Define the metadata extraction function.
def metadata_func(record: dict, metadata: dict) -> dict:
    for k in record:

        metadata["sessionuid"] = record.get("sessionuid")
        metadata["speakers"] = record.get("speakers")
        metadata["location"] = record.get("location")

    if "source" in metadata:
        source = metadata["source"].split("/")
        source = source[source.index("Users"):]
        metadata["source" ] = "/".join(source)

    return metadata
```

```python
loaders = []
for n in range(1606):
    session = f'data_json/element_{n}.json'
    loaders.append(JSONLoader(file_path=session,jq_schema='.',
    content_key='summary', metadata_func=metadata_func))
```

## Embeddings
https://github.com/aws-samples/rag-using-langchain-amazon-bedrock-and-opensearch/blob/main/ask-bedrock-with-rag.py

```python
import boto3
from langchain.embeddings import BedrockEmbeddings

bedrock_client = boto3.client("bedrock-runtime", region_name="us-east-1")

bedrock_embedding_model_id = 'amazon.titan-embed-text-v1'
```

```python
def create_langchain_vector_embedding_using_bedrock(bedrock_client, bedrock_embedding_model_id):
    bedrock_embeddings_client = BedrockEmbeddings(
        client=bedrock_client,
        model_id=bedrock_embedding_model_id)
    return bedrock_embeddings_client
```

```python
bedrock_embeddings_client = create_langchain_vector_embedding_using_bedrock(bedrock_client, bedrock_embedding_model_id)
```

### Split

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size = 1000,
    chunk_overlap = 150
)
```

```python
splits = []
for doc in docs:
    splits.append(text_splitter.split_documents(doc))
```
